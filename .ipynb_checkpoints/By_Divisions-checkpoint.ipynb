```python
import pandas as pd
import numpy as np
import scipy as sp
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/clean_stop_data.csv')
```

```python
df.head()
```

```python
by_division = df.groupby(['division', 'officer_race', 'citizen_race']).aggregate(np.sum)[['searched', 'frisked', 'citation_issued']]
```

```python
by_division['stops'] = df.groupby(['division', 'officer_race', 'citizen_race'])['year'].count()
```

```python
by_division.drop(['NON LMPD', '8', 'STM', '7', 'SHV', '6', '4', '5', '1', '3', 'JTN', 'ANC'], inplace=True)
```

```python
by_division
```

```python
by_division['% searched'] = by_division['searched'] / by_division['stops']
by_division['% cited'] = by_division['citation_issued'] / by_division['stops']
by_division['% of all stops'] = by_division['stops'] / df.year.count()
by_division['% of division stops'] = by_division['stops'] / by_division.groupby('division').aggregate(np.sum)['stops']
```

```python
def plot_beta(alpha, beta, ax, dist, title=None, label=None):

    x = np.linspace(0.0, 1.0, beta)
    y = dist.pdf(x)    
    
    lines = ax.plot(x, y, label=label)
    ax.fill_between(x, y, alpha=0.2, color=lines[0].get_c())
    if title: 
        ax.set_title(title)
    ax.set_ylim(0.0, np.max(y)*1.2)   
```

```python
def a_b_division(race, division):
    a = df.groupby(['division', 'citizen_race']).sum('searched').loc[division].loc[race, 'searched'] # alpha
    b = df.groupby(['division', 'citizen_race']).count().loc[division].loc[race, 'searched'] # beta
    return stats.beta((a + 1), b - a + 1), a, b
```
